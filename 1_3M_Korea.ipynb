```python
import pandas as pd
from google.colab import files

# Step 1: Upload the file in Colab


# Load the dataset
try:
    df = pd.read_csv('/content/1.3M-Korea-ggumim.co.kr-Decorating-the-House-UsersDB-csv-2020.csv', encoding='utf-8-sig')
    print("Dataset loaded successfully.")
except Exception as e:
    print(f'Error loading dataset: {e}')
    raise

# Step 2: Validate email addresses
# Define a regex pattern for email validation
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

# Create a mask for valid emails
valid_email_mask = df['email'].str.match(email_pattern, na=False)

# Step 3: Convert `created_date_time` to proper datetime format (if present)
if 'created_date_time' in df.columns:
    df['created_date_time'] = pd.to_datetime(df['created_date_time'], errors='coerce')
    print("Converted 'created_date_time' to datetime format.")
    # Mark rows where the conversion failed as junk
    invalid_date_mask = df['created_date_time'].isna()
else:
    invalid_date_mask = pd.Series([False] * len(df))  # No invalid dates if column doesn't exist

# Step 4: Identify duplicate rows
duplicate_mask = df.duplicated()

# Step 5: Separate valid and junk datasets
valid_mask = valid_email_mask & ~invalid_date_mask & ~duplicate_mask
cleaned_df = df[valid_mask].copy()
junk_df = df[~valid_mask].copy()

print(f"Number of valid rows: {len(cleaned_df)}")
print(f"Number of junk rows: {len(junk_df)} (includes invalid emails, bad dates, and duplicates)")

# Step 6: Save the cleaned and junk datasets
cleaned_filename = 'cleaned_dataset.csv'
junk_filename = 'junk_dataset.csv'

try:
    cleaned_df.to_csv(cleaned_filename, index=False)
    junk_df.to_csv(junk_filename, index=False)
    print(f"Cleaned dataset saved to '{cleaned_filename}'.")
    print(f"Junk dataset saved to '{junk_filename}'.")
    # Automatically download the files

except Exception as e:
    print(f"Error saving datasets: {e}")
    raise
```
